```python
import lightkurve as lk

tpf = lk.search_tesscut('TIC 120459871').download(cutout_size=20)


searchresult=tpf.SearchResult
tpf.plot()
```

```python
import numpy as np
aper = np.zeros(tpf.shape[1:], dtype=np.int)
aper[10:13, 8:11] = 1
aper_background = np.zeros(tpf.shape[1:], dtype=np.int)
aper_background[15:18, 8:11] = 1
tpf.plot(aperture_mask=aper, mask_color='red');
tpf.plot(aperture_mask=aper_background, mask_color='green')
```

```python
tpf.interpret()
```

```python
import matplotlib.pyplot as plt


# Build the NEW aperture, and the light curve
user_lc = tpf.to_lightcurve(aperture_mask=aper.astype(bool))
# Clean the light curve
user_lc = user_lc.remove_nans().remove_outliers()
user_lc.plot(marker='o',linestyle='None',markersize=4,color='blue')
plt.ylim(0.99,1.01)

# Build the BACKGROUND aperture, and the light curve
background_lc = tpf.to_lightcurve(aperture_mask=aper_background.astype(bool))
# Clean the light curve
background_lc = background_lc.remove_nans().remove_outliers()
background_lc.plot(marker='o',linestyle='None',markersize=4,color='red')
plt.ylim(0.99,1.01)

#total_lc = user_lc/background_lc
#total_lc.plot(marker='o',linestyle='None',markersize=4,color='green')
```

```python
#SFF Corrector
corrector = lk.correctors.SFFCorrector(user_lc)
new_lc_sff = corrector.correct(user_lc.centroid_col,user_lc.centroid_row)

new_lc_sff.plot()
```

```python
#PLD Corrector
corrector = lk.PLDCorrector(tpf) # doctest: +SKIP
new_lc_pld = corrector.correct() # doctest: +SKIP

new_lc_pld.plot()
```

```python
user_lc.scatter()
```

```python
#SFF periodogram
pg_sff = new_lc_sff.to_periodogram(oversample_factor=1)
pg_sff.plot()
```

```python
pg_sff.plot(view='period', scale='log')
```

```python
period = pg_sff.period_at_max_power
print('Best period: {}'.format(period))
new_lc_sff.fold(period.value).scatter();
```

```python
period = pg_sff.period_at_max_power * 4
print('Best period: {}'.format(period))
new_lc_sff.fold(period.value).scatter();
```

```python
import astropy.units as u
pg_sff = new_lc_sff.to_periodogram(minimum_period=0.9*u.day, maximum_period=1.2*u.day, oversample_factor=10)
```

```python
period = pg_sff.period_at_max_power
print('Best period: {}'.format(period))
new_lc_sff.fold(period.value).scatter();
```

```python
#PLD Periodogram
pg_pld = new_lc_pld.to_periodogram(oversample_factor=1)
pg_pld.plot()
```

```python
pg_pld.plot(view='period', scale='log')
```

```python
period = pg_pld.period_at_max_power
print('Best period: {}'.format(period))
new_lc_pld.fold(period.value).scatter();
```
